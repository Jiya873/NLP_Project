```python
import pandas as pd

# Read the CSV file
df = pd.read_csv("Replika_reviews.csv")

# Display the first few rows of the dataframe
df.head()
```

```python
df.drop(columns=['source', 'user_name', 'review_title'], inplace=True)
df.head()
```

```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
%pip install textblob
from textblob import TextBlob

# Function to detect language using TextBlob
def detect_language(comment):
    try:
        return TextBlob(comment).detect_language()
    except:
        return None

# Apply the function to the review_description column
df['detected_language'] = df['review_description'].apply(detect_language)

# Display the dataframe with the new column
df.head()
```

```python
%pip install langdetect

from langdetect import detect

# Assuming df is your DataFrame
def is_proper_english(comment):
    try:
        language = detect(comment)
        return language == 'en'
    except:
        return False

# Apply the function to the review_description column
df['is_proper_english'] = df['review_description'].apply(is_proper_english)

# Filter out comments that are not in proper English
non_english_comments = df[~df['is_proper_english']]

# Display the non-English comments
print(non_english_comments[['review_id', 'review_description', 'is_proper_english']])
```
