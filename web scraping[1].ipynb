```python
from google_play_scraper import app, Sort, reviews_all
from app_store_scraper import AppStore
import pandas as pd
import numpy as np
import json
import os
import uuid
import time
import random
```

```python
def fetch_google_play_reviews(app_id, max_retries=5):
    retries = 0
    while retries < max_retries:
        try:
            g_reviews = reviews_all(
                app_id,
                sleep_milliseconds=1000,  # 1 second delay between requests
                lang='en',
                country='us',
                sort=Sort.NEWEST,
            )
            return g_reviews
        except Exception as e:
            print(f"Error fetching Google Play reviews: {e}")
            sleep_time = (2 ** retries) + random.uniform(0, 1)
            print(f"Retrying in {sleep_time:.2f} seconds")
            time.sleep(sleep_time)
            retries += 1
    print("Max retries reached. Unable to fetch Google Play reviews.")
    return None

google_play_reviews = fetch_google_play_reviews("app_id")
google_play_reviews
```

```python
g_df = pd.DataFrame(np.array(google_play_reviews), columns=['review'])
g_df2 = g_df.join(pd.DataFrame(g_df.pop('review').tolist()))

g_df2
```

```python
g_df2.rename(columns= {'score': 'rating','userName': 'user_name', 'reviewId': 'review_id', 'content': 'review_description', 'at': 'review_date', 'replyContent': 'developer_response', 'repliedAt': 'developer_response_date', 'thumbsUpCount': 'thumbs_up'},inplace = True)
```

```python
def fetch_app_store_reviews(app_name, app_id, max_retries=5):
    retries = 0
    while retries < max_retries:
        try:
            app = AppStore(country="in", app_name=app_name, app_id=app_id)
            app.review()
            return app.reviews
        except Exception as e:
            print(f"Error fetching App Store reviews: {e}")
            sleep_time = (2 ** retries) + random.uniform(0, 1)
            print(f"Retrying in {sleep_time:.2f} seconds")
            time.sleep(sleep_time)
            retries += 1
    print("Max retries reached. Unable to fetch App Store reviews.")
    return None

app_store_reviews = fetch_app_store_reviews("app", "app_id")
print(app_store_reviews)
```

```python
app_store_df = pd.DataFrame(app_store_reviews)
```
