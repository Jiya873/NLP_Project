```python
from google_play_scraper import app, Sort, reviews_all
from app_store_scraper import AppStore
import pandas as pd
import numpy as np
import json
import os
import uuid
import time
import random

def fetch_google_play_reviews(app_id, max_retries=5):
    retries = 0
    while retries < max_retries:
        try:
            g_reviews = reviews_all(
                app_id,
                sleep_milliseconds=1000,  # 1 second delay between requests
                lang='en',
                country='us',
                sort=Sort.NEWEST,
            )
            return g_reviews
        except Exception as e:
            print(f"Error fetching Google Play reviews: {e}")
            sleep_time = (2 ** retries) + random.uniform(0, 1)
            print(f"Retrying in {sleep_time:.2f} seconds")
            time.sleep(sleep_time)
            retries += 1
    print("Max retries reached. Unable to fetch Google Play reviews.")
    return None

google_play_reviews = fetch_google_play_reviews("ai.replika.app")
google_play_reviews
```

```python
from time import sleep
from selenium import webdriver
from bs4 import BeautifulSoup as soup
import pandas as pd
#chromedriver
browser = webdriver.Chrome("C:\Windows\chromedriver.exe")

#google playstore app id url here
app_url_id ="https://play.google.com/store/apps/details?id=ai.replika.app&hl=en"
review_list = []

def scrape_reviews():
     browser.get(app_url_id+"&showAllReviews=true")
     sleep(3)
     #navigate to BeautifulSoup
     source = soup(browser.page_source,"html.parser")
     user_data = source.find_all("div", class_="EGFGHd")
     for items in user_data:
        d={}
        d["names"] = items.find("span", {"class":"X5PpBb"}).text
        d["rating"] = items.find("div", class_="iXRFPc").find("div")["aria-label"]
        d["date"] = items.find("span", class_="bp9Aid").text

        try:
         d["reviews"]= items.find("div", class_="h3YV2d").find("span").text
        except:
         d["reviews"] = None
         review_list.append(d)

scrape_reviews()
#remove duplicates
for n in review_list:
  for k,v in n.items():
     if v == None:
        review_list.remove(n)

df1 = pd.DataFrame(review_list)
```
